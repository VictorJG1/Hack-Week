# Number Randomizer

```python
import random
 
list = [3, 2, 4, 1, 67, 45]
 
print("List before being random : ", end="")
for r in range(0, len(list)):
    print(list[r], end=" ")
print("\r")

random.shuffle(list)
 
print("The list is after being random from the original : ", end="")
for r in range(0, len(list)):
    print(list[r], end=" ")
print("\r")

print("point average of 100-1000: ", end="")
print(random.uniform(100, 1000))
```
